## Part 1 — What is a Website?

Every webpage you visit in a browser is built from a text file called **HTML** (HyperText Markup Language). HTML uses *tags* — words wrapped in angle brackets — to describe what each piece of content is.

```html
<h3>Alex Rivera</h3>         ← a heading
<p>Digital Studies</p>       ← a paragraph
```

A browser reads those tags and turns them into the page you see. The *content* (names, text, links) and the *instructions* (tags) are mixed together in the same file.

Generally, HTML handles the structure of the page. As you will learn later, CSS controls styling and JavaScript controls the interactivity. These three languages, HTML, CSS, and JS are the foundation of modern web coding.

## Part 4 — The Problem with Hardcoding

What you just did is called **hardcoding** — writing specific values directly into a file that is supposed to be *structure*, not *content*.

Here's the core problem:

> When data and presentation are mixed together, every change to the data requires a human to find and edit code.

In a small team this is annoying. At real scale — a company website with thousands of employees, a university directory, a news article database — it becomes impossible.

The solution is **separation of concerns**: keep the data in one place, keep the presentation instructions in another, and write code that connects them.

You already use this pattern every day without thinking about it. A Google Sheets spreadsheet is *data*. The chart Google Sheets draws from it is *presentation*. If you add a row to the spreadsheet, the chart updates automatically. You didn't touch the chart.

That is exactly what we are going to do with your team page.

## Part 5 — Data in a Spreadsheet

**What is a CSV, and why use it?**

A **CSV** (Comma-Separated Values) file is the simplest possible spreadsheet format. Think of it as the spreadsheet version of a **.txt** file. It does not allow fancy formatting like an Excel file (.xls) or a Google Sheet (.gsheet), but since it is just raw data it is highly compatible. Pretty much any spreadsheet software can open CSV files, which makes them easy to share and process with code.

## Part 6 — What is a Variable?

Before we run any code, one concept to understand: a **variable** is just a label the computer uses to remember a value.

```python
name = "Alex Rivera"    # the label is 'name'; the value is 'Alex Rivera'
major = "Digital Studies"
```

You've been doing this in HTML too — just without Python syntax. The difference is that in Python, variables can be **computed** from other variables, and they can be **reused** anywhere in the program without retyping.

When we load your `team.csv` into Python, every column header becomes a variable Python can work with. Every row becomes a set of values for those variables. Python can then use those variables to write the JSON file that the team page reads.

## Part 8 — What Just Happened?

Open `generate_team_json.py` (<img src="../lesson_assets/images/vscode/file-text.svg" alt="File icon" width="14">) in **Explorer** (<img src="../lesson_assets/images/vscode/files.svg" alt="Explorer icon" width="14">) and look at these three key lines:

```python
df = pd.read_csv(INPUT_CSV)          # Input: load the spreadsheet
df = df.sort_values("_last_name")    # Compute: sort the rows
json.dump(output, f)                 # Output: write the JSON file
```

These three lines — **read, transform, write** — are the fundamental pattern of almost everything we will do in this course:

| Step | Name | What it means |
|---|---|---|
| Read | **Ingest** | Get data from somewhere into Python |
| Transform | **Wrangling / Analysis** | Shape or analyze the data |
| Write | **Output** | Save a result or build a presentation |

In future lessons the data will be thousands of Reddit posts, and the transform step will involve finding locations and measuring sentiment. But the pattern — and Python’s role in it — is exactly the same.
