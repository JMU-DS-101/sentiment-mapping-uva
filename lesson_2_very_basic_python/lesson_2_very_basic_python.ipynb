### What Are Variables?

Adding specific numbers like `2 + 1` is fine, but the real magic happens when we use **variables**.

**Think of a variable like a labeled box** where you can store information. 

In the example below:
- We put the number `2` in a box labeled `x`  
- Then we can use `x` instead of writing `2` again
- `x + 1` gives us the same result as `2 + 1`

**Why is this useful?** You'll see in a moment! 🎯

````python
x = 2
x + 1
````

### How Variables Work

The `=` sign in Python means **"store this value"** - it's like putting something in a labeled box.

```python
x = 2
```

**Read this as:** "Store the number 2 in a box labeled x"

⚠️ **Important:** In Python, `=` does NOT mean "equals" like in math class. It means "store" or "assign."

### Why Variables

It may not be entirely obvious why you would want to store things in a variable. If you want to add `2` + `1`, storing `2` in the variable `x` seems like an extra step. The reason for doing so is that you do not always know what the information is going to be. 

**Example: Calculating Area of a Rectangle**

Imagine you need to calculate the area of different rectangles. Without variables, you'd have to write:
- Room 1: `12 * 8 = 96`  
- Room 2: `15 * 10 = 150`
- Room 3: `9 * 12 = 108`

But with variables, you can write the formula once and reuse it:

```python
length = 12
width = 8
area = length * width
```

Now you can easily change the `length` and `width` values to calculate any rectangle's area. This becomes incredibly powerful when you're processing hundreds of rooms in a building or thousands of data points!

Two additional types you will encounter:

- **Boolean** (`bool`) — True or False values, used in logic

Python automatically figures out what type of data you're storing. This makes Python easier to use than many other programming languages!

````python
# Integer no decimal
a = 1

#float decimal
b = 2.0

#String quote marks
c = "Three"

#Bool reserved word True or False
d = True
````

### Numbers vs "Numbers"
Sometimes numbers look like numbers, but they're actually text! Can you spot the problem below?

````python
number1 = 567
number2 = "5309"
number1 + number2
````

**Problem Alert!** ❌

This gives us a TypeError because:
- `567` is a number (integer)  
- `"5309"` is text (string) - notice the quotes!

Even though "5309" looks like a number, the quotes make Python treat it as text. You can't do math with text!

### Type Conversion 🔄

**Good news!** You can convert between data types using built-in functions:

- `int()` → Convert to whole number: `int("5309")` → `5309`
- `float()` → Convert to decimal: `float("3.14")` → `3.14`
- `str()` → Convert to text: `str(567)` → `"567"`
- `bool()` → Convert to True/False

**How to use:** Put your variable inside the parentheses: `int("5309")`

**Now our math works!** ✅ By converting the string to an integer with `int()`, we can add them:

```python
number1 + int(number2)   # int() converts "5309" → 5309
```

````python
number1 = 567
number2 = "5309"
number1 + int(number2)
````

## Part 2: Calling Functions & Methods

Variables store data. **Functions and methods** are how you do things with that data. In these lessons you will mostly be *calling* functions that have already been written for you — not writing your own.

### Calling Functions

You have already been using **functions** — `type()`, `print()`, and `int()` are all functions. A function is a named operation: you give it input inside parentheses, and it produces output.

```python
type(a)           # tells you the data type of a
print("Hello")    # displays text
int("5309")       # converts the string "5309" to the integer 5309
```

Some functions take **keyword arguments** — named options that control exactly how the function runs. You will see this pattern constantly:

```python
pd.read_csv("data.csv", encoding="utf-8")
df.sample(n=5, random_state=42)
```

The `n=5` and `encoding="utf-8"` parts are keyword arguments. The `name=value` pattern is all you need to recognize — the documentation for each function will always tell you what names are available.

### Methods

A **method** is a function that belongs to a specific data type and is accessed with dot notation. When you write `my_string.lower()`, you are asking the string to perform an operation on itself.

The key pattern to recognize is:

```
variable.method_name(arguments)
```

In Lesson 3 you will start working with data tables (called DataFrames in pandas). Methods appear there too — they work the same way, just applied to columns instead of individual strings:

```python
df['university'].str.lower()    # convert a whole column to lowercase
df['score'].mean()              # calculate the average of a column
```

### Task 4: Predict and Run

String methods follow very consistent naming conventions. Look at the two examples in the code cell below — they run `.lower()` and `.upper()` on the same string.

**Before running the cell:** predict what the third line — `name.title()` — will produce.

Then run the cell to confirm.

````python
name = "james madison university"

print(name.lower())   # → "james madison university"
print(name.upper())   # → "JAMES MADISON UNIVERSITY"
print(name.title())   # → ???
````

**Common string methods — keep this as a reference:**

| Method | What it does | Example |
|---|---|---|
| `.lower()` | all lowercase | `"JMU".lower()` → `"jmu"` |
| `.upper()` | ALL UPPERCASE | `"jmu".upper()` → `"JMU"` |
| `.title()` | Title Case | `"james madison".title()` → `"James Madison"` |
| `.strip()` | remove leading/trailing spaces | `"  JMU  ".strip()` → `"JMU"` |
| `.replace("old", "new")` | swap text | `"JMU".replace("JMU", "Dukes")` → `"Dukes"` |

Methods always use **dot notation**: the variable, a dot, the method name, and parentheses.

You can also **chain methods** — call one after another — in a single line:

```python
messy_text = "  COLONIAL WILLIAMSBURG  "
clean_text = messy_text.strip().title()
# Result: "Colonial Williamsburg"
```

This is common in data cleaning: `.strip()` removes extra spaces, then `.title()` capitalizes properly. You will see chains like this throughout these lessons.

## Part 3: Packages

In these lessons you will be working with libraries — collections of pre-built functions — rather than writing your own. Understanding how to load and use a package is one of the most practical Python skills you can have.

**What is a package?**

A **package** is a collection of pre-written functions that other people have built and shared. Instead of writing every tool from scratch, you import a package and use the tools inside it.

The `import` statement loads a package. The `as` keyword gives it a short nickname so you don't have to type the full name each time:

```python
import pandas as pd       # data tables — nicknamed "pd"
import nltk               # natural language processing
import spacy              # location and entity extraction
```

Packages are loaded once at the top of a notebook and then available everywhere below.

````python
import pandas as pd
````

Once a package is imported, you access its functions using the prefix:

```python
pd.read_csv("data.csv")       # load a CSV file
pd.DataFrame(my_list)         # convert a list into a table
```

You can also pass keyword arguments (the `name=value` pattern from Part 2) to customize how functions run:

```python
df = pd.read_csv("data.csv", encoding="utf-8")
df.head(10)
```

You do not need to memorize which keyword arguments exist — the documentation (or a quick search) will always tell you what options are available for a given function.

## Part 4: Reading Code

As you work through these lessons you will encounter Python patterns that you do not need to write yourself, but that you do need to understand. This section gives you just enough background to read that code confidently.

### For Loops

One of the most common features of any programming language is **iteration** — repeating the same operation for every item in a collection. In Python this is done with a `for` loop.

A `for` loop has two parts:
- A **header** that says what to loop over
- An **indented body** that runs once for each item

The loop below goes through every university in a list and prints it in uppercase. **Read it as:** "For each `x` in `list4`, print `x` in uppercase."

You will see this pattern throughout these lessons. You do not need to write `for` loops — just recognize what they are doing when you see them.

````python
list4 = ["jmu", "uva", "virginia tech", "william and mary"]

for x in list4:
    print(x.upper())
````

### Lambda

A `lambda` is a compact, one-line function. You will see it most often with `.apply()`, which runs an operation on every row of a data column:

```python
df['score'].apply(lambda x: x * 100)
```

**How to read it:** "For each value `x` in the `score` column, multiply it by 100."

Think of `lambda x:` as "a temporary variable called `x` that holds each item one at a time." You will not write lambdas in these lessons, but you will see them regularly.

```python
# Another common example: classifying sentiment scores
df['sentiment'].apply(lambda x: "positive" if x > 0 else "negative")
```

### Try/Except

A `try/except` block handles errors gracefully. Instead of crashing, Python tries one thing and falls back to another if it fails:

```python
try:
    result = int(value)
except:
    result = 0
```

**How to read it:** "Try to convert `value` to an integer. If that causes an error, set `result` to `0` instead."

This is common in data cleaning when some values in a column cannot be converted — rather than crashing the whole script, `try/except` lets the code handle bad data and keep running.
